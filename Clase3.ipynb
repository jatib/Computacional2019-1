```python
from pylab import *
from matplotlib import *
```

```python
x = arange(-1,10,0.01)
```

```python
plot(x,cos(x),'--m');axhline(y=0,color='r',linestyle='--');grid(True);title('Coseno');xlabel("Theta");ylabel("Y");
```

```python
def incresearch(f,x,d,limite=100):
    while f(x)*f(x+d) > 0:
        if x > limite: return "Nada"
        x = x+d
    return [f(x),f(x+d),x]
```

```python
def fun(x):
    return x**3-10*x**2+5
```

```python
a = incresearch(fun,-1,1e-2)
print(a)
```

```python
plot(x,fun(x),'--m',a[2],a[0],"go");axhline(y=0,color='r',linestyle='--');grid(True);title('Función nueva');xlabel("x");ylabel("y");
```
